```python
import rpy2.robjects as robjects
from rpy2.robjects.conversion import localconverter
from rpy2.robjects import pandas2ri

import pandas as pd
```

```python
pd.set_option('display.max_columns', None)
```

```python
readRDS = robjects.r['readRDS']
```

```python
survey_r = readRDS(r"C:\Users\Peter\OneDrive\Desktop\Code and Projects\ECO1400\Replication Package\Datasets\Survey\survey.rds")
like_user_r = readRDS(r"C:\Users\Peter\OneDrive\Desktop\Code and Projects\ECO1400\Replication Package\Datasets\Likes\likes_user.rds")
fb_user_r = readRDS(r"C:\Users\Peter\OneDrive\Desktop\Code and Projects\ECO1400\Replication Package\Datasets\Extension\facebook\fb_user.rds")
nav_his_user_r = readRDS(r"C:\Users\Peter\OneDrive\Desktop\Code and Projects\ECO1400\Replication Package\Datasets\Extension\navigate/nav_user.rds")
shared_user_r = readRDS(r"C:\Users\Peter\OneDrive\Desktop\Code and Projects\ECO1400\Replication Package\Datasets\Shared\shared_user.rds")
```

```python
survey_r
```

```python
with localconverter(robjects.default_converter + pandas2ri.converter):
    survey = robjects.conversion.rpy2py(survey_r)
    like_user = robjects.conversion.rpy2py(like_user_r)
    fb_user = robjects.conversion.rpy2py(fb_user_r)
    nav_his_user = robjects.conversion.rpy2py(nav_his_user_r)
    shared_user = robjects.conversion.rpy2py(shared_user_r)
```
